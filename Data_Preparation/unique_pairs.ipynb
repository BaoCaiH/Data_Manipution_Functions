```python
def unique_pairs(df, column_1, column_2):
    '''
    Return a pandas dataframe which index is the unique pairs of the 2 columns and the values are the frequencies
    Argument:
        df: a data frame
        column_1: first column in the pair
        column_2: second column in the pair
        
    '''
    import pandas as pd
    return df.loc[:, [column_1, column_2, df.columns[0]]].groupby([column_1, column_2]).\
            count().rename({df.columns[0] : 'frequency'}, axis = 1)
```
